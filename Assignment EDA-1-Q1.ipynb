# Convert the categorical columns using encoder.

```python
# use of LabelEncoder
```

```python
df_cat = df.select_dtypes("object")
df_cat
```

```python
df_num = df.select_dtypes(["int","float"])
df_num
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in df_cat:
    df_cat[i] = le.fit_transform(df_cat[i])
df_cat
```

```python
new_num = pd.concat([df_cat,df_num],axis = 1) # Concatenation happens column vise
new_num
```
