```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = "Cabin", axis = True, inplace = True)
df.fillna(df["Age"].mean(), inplace = True)
df
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.drop(["PassengerId","Name", "Ticket", "Embarked"], axis = 1, inplace = True)
```

```python
df
```

```python
df["Family"] = df["SibSp"] + df["Parch"]
```

```python
df.drop(["SibSp","Parch"], axis = 1, inplace = True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Sex"] = le.fit_transform(df["Sex"])
df
```

```python
x = df.iloc[:,1:]
x
```

```python
y = df.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
svm = SVC()
```

```python
def mymodel(model):
    model.fit(xtrain, ytrain)
    ypred = model.predict(xtest)
    print(classification_report(ytest, ypred))
    return model
```

```python
mymodel(lr)
```

```python
mymodel(knn)
```

```python
mymodel(dt)
```

```python
mymodel(svm)
```

```python
svm = SVC(kernel = "linear")
mymodel(svm)
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
param_grid = {"C":[1,10], "gamma": [1,10]}
gs = GridSearchCV(svm, param_grid, verbose = 4)
gs.fit(xtrain,ytrain)
```

```python
gs.best_params_
```

```python
svm = SVC(kernel = "linear", C = 1, gamma = 1)
mymodel(svm)
```
