# Machine Learning

```python
data = pd.read_csv("Scores.csv")
data
```

```python
# Separation of x and y.
x = data.iloc[:,:-1]
x
```

```python
y = data.iloc[:,-1]
y
```

```python
# Train Test Split the Data
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest,  = train_test_split(x,y, test_size = 0.3, random_state = 1)
```

```python
xtrain
```

```python
ytrain
```

```python
# LinearRegression Model
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(xtrain,ytrain)

ypred = lr.predict(xtest)

print(lr.coef_) # Slope or Gradient or Coeficient

print(lr.intercept_) # Y-intercept
```

```python
# Model performing on the Training Set
plt.scatter(xtrain, ytrain, color ="red")
plt.plot(xtrain, lr.predict(xtrain), color = "blue")
plt.grid()
plt.show()
```

```python
# Model performing on the Testing Set 
plt.scatter(xtest, ytest, color = "red")
plt.plot(xtest, lr.predict(xtest), color = "blue")
plt.grid()
plt.show()
```

```python
# Model Evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(ytest,ypred)
mse = mean_squared_error(ytest,ypred)
rmse = np.sqrt(mse)
r2 = r2_score(ytest,ypred)
print(f"MAE {mae}\nMSE {mse}\nRMSE {rmse}\nAccuracy {r2}")
```

# Regularization (RIDGE and LASSO)

```python
# After doing EDA and Preprocessing and before Machine Learnig Model preparation we seperate the values for X and Y(Target Column).
```

```python
x = df_new.iloc[:, :-1] 
x
```

```python
y = df_new.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.3, random_state = 1)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain, ytrain)
```

```python
train = lr.score(xtrain, ytrain)
test = lr.score(xtest, ytest)
print(f"Training Accuracy is {train}")
print(f"Testing Accuracy is {test}")
```

```python
'''
We can see that Training Accuracy is more than Testing Accuracy which means that Training error (Bias) gonna be less and We can 
also see that Testing Accuracy is less than Training Accuracy which means that Testing error (Variance) gonna be more that's why
Testing Accuracy is less because of more Testing error (Variance) so that's why we can say that the model is overfit and we have  
to do regularization in order to reduce the overfitting, the types of regularization are L2(RIDGE: Regularization Regression) and
L1(LASSO: Least Absolute Shrinkage and Selection Operator).
'''
```

# 1) L2 Regularization (RIDGE: Regularization Regression)  

```python
'''
Working of L2: The slope(m1,m2) of features with bad correlation with target column get reduced(not upto to zero) so that the 
overall slope(m) of all the features get reduced and Y-Intercept also reduces so that model which has a overfit line get tilt 
and become best fit line and model become good fit.
'''
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
l2 = Ridge(alpha = 5)
l2.fit(xtrain,ytrain)
```

```python
train = l2.score(xtrain, ytrain)
test = l2.score(xtest, ytest)
print(f"Training Accuracy is {train}")
print(f"Testing Accuracy is {test}")
```

```python
#hypertuning lambda/ alpha value: It is used to have the control on the fact that how much error we want to put in the data.
for i in range(1,30):
    l2 = Ridge(alpha = i)
    l2.fit(xtrain, ytrain)
    test = l2.score(xtest, ytest)
    print(f"For Value of Lambda {i} Testing Accuracy is {test}")
```

```python
''' 
We find that we get best Testing Accuracy for lambda/ alpha = 1, after that Testing Accuracy decreases so alpha = 1 is threshold
value for alpha, So that's why using alpha = 1 in below cell.
'''
```

```python
l2 = Ridge(alpha = 1)
l2.fit(xtrain, ytrain)
```

```python
train = l2.score(xtrain, ytrain)
test = l2.score(xtest, ytest)
print(f"Training Accuracy is {train}") # Best possible Trainig Accuracy we can get from Ridge after hypertuning lambda. 
print(f"Testing Accuracy is {test}") # Best possible Testing Accuracy we can get from Ridge after hypertuning lambda.
```

# 2) L1 Regularization (LASSO : Least Absolute Shrinkage and Selection Operator)

```python
'''
Working of L1: The slope(m1,m2) of features with bad correlation with target column become zero by deleting that features(dimensionality reduction) 
so that the overall slope(m) of all the features get reduced and Y-Intercept also reduces so that model which has a overfit 
line get tilt and become best fit line and model become good fit. 
'''
```

```python
for i in range(1,50):
    l1 = Lasso(alpha = i)
    l1.fit(xtrain, ytrain)
    test = l1.score(xtest, ytest)
    print(f"For Value of Lambda {i} Testing Accuracy is {test}")
```

```python
''' 
We find that we get best Testing Accuracy for lambda/ alpha = 1, after that Testing Accuracy decreases so alpha = 1 is threshold
value for alpha, So that's why using alpha = 1 in below cell.
'''
```

```python
l1 = Lasso(alpha = 1)
l1.fit(xtrain, ytrain)
train = l1.score(xtrain, ytrain)
test = l1.score(xtest, ytest)
print(f"Training Accuracy is {train}")
print(f"Testing Accuracy is {test}")
```

```python
'''
We can see that both(Ridge and Lasso) has nearly same amount of difference between Training Accuracy and Testing Accuracy, that's 
why we can use any of them or we need to use l1 and l2 according to organization needs if they tells you that you can remove less 
relevant features/column then you can use Lasso(L1) in which you actually do dimensionality reduction(reduction of feature/column)
and if they tell to use all the features then you need to use Ridge(L2) in which you don't do dimensionality reduction.
'''
```
